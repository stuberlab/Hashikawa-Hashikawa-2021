```r
MPOA_Harmony<-merge(x = P23M, y = P35M)
MPOA_Harmony<-merge(x = MPOA_Harmony, y = AM)
MPOA_Harmony<-merge(x = MPOA_Harmony, y = Cast)
MPOA_Harmony<-merge(x = MPOA_Harmony, y = P23F)
MPOA_Harmony<-merge(x = MPOA_Harmony, y = P35F)
MPOA_Harmony<-merge(x = MPOA_Harmony, y = AF)
MPOA_Harmony<-merge(x = MPOA_Harmony, y = OVX)
```

```r
MPOA_Harmony<- NormalizeData(object =MPOA_Harmony,verbose = FALSE) 
```

```r
MPOA_Harmony<-FindVariableFeatures(object =MPOA_Harmony,selection.method = "vst", nfeatures = 2000, verbose = FALSE)
```

```r
MPOA_Harmony<- ScaleData(object = MPOA_Harmony, verbose = FALSE)
```

```r
MPOA_Harmony<- RunPCA(object = MPOA_Harmony, features = VariableFeatures(object =MPOA_Harmony), verbose = FALSE)
MPOA_Harmony <-  RunHarmony(MPOA_Harmony,"stim")
MPOA_Harmony <- RunUMAP(MPOA_Harmony,reduction = "harmony", dims = 1:40) 
MPOA_Harmony <-FindNeighbors(MPOA_Harmony,reduction = "harmony", dims = 1:40)
MPOA_Harmony <-FindClusters(MPOA_Harmony,resolution = 0.8, dims = 1:40)
```

```r
#diversity
diversity_index<-numeric()
diversity_index<-lapply(1:dim(MPOA.integrated@meta.data)[1], function(i) {
   X<-MPOA.integrated@reductions$umap@cell.embeddings[,"UMAP_1"] -MPOA.integrated@reductions$umap@cell.embeddings[,"UMAP_1"][i]
   Y<-MPOA.integrated@reductions$umap@cell.embeddings[,"UMAP_2"] -MPOA.integrated@reductions$umap@cell.embeddings[,"UMAP_2"][i]
   D<-(X^2+Y^2)^0.5
    neighbor<-MPOA.integrated@meta.data$stim[order(D)[1:8]]
    length(unique(neighbor))
   })
```

```r
MPOA.integrated@meta.data$diversity_index<-as.numeric(diversity_index)
```
